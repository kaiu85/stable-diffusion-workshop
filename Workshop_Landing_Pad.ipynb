The reason that I took these examples from the twitter feed of a (back-then) Google researcher, instead of creating some images by my own, was the following: In summer 2021, i.e. just about one year ago, it just wasn't possible to run your own diffusion models on private hardware (i.e., gaming-level PCs) or on free Colab instances. This all changed this year, when a group of researchers from Munich (but originally from Heidelberg) published "Stable Diffusion", which comes close to performance to the large models developed by Google and OpenAI, but can be run on graphics cards with about 11 GB of video memory.

This puts us in the position, to actually be able to do some hands on work with state-of-the art diffusion models today, just here in your browser, using Google Colab. This is a free, web-based service, which allows you to generate notebooks **just like this one here**, which can include text, images, videos, but also Python code, you can execute directly in your browser. In reality, it will be executed on a cloud based machine, of course. In many aspects, these machines are similar to a desktop pc, in that they can provide you with 12 GB of RAM and a graphics card with about 16GB of video memory.
